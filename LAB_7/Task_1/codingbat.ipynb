### Warmup-1

```python
def sleep_in(weekday, vacation):
  if not weekday or vacation:
    return True
  else:
    return False

print(sleep_in(True,False))
```

```python
def diff21(n):
  if n <= 21:
    return 21 - n
  else:
    return (n - 21) * 2

print(diff21(15))
```

```python
def near_hundred(n):
  return ((abs(100 - n) <= 10) or (abs(200 - n) <= 10))

print(near_hundred(120))
```

```python
def missing_char(str, n):
  front = str[:n]   # up to but not including n
  back = str[n+1:]  # n+1 through end of string
  return front + back

print(missing_char("salem",4))
```

```python
def monkey_trouble(a_smile, b_smile):
  if a_smile and b_smile:
    return True
  if not a_smile and not b_smile:
    return True
  return False

print(monkey_trouble(True,True))
```

```python
def parrot_trouble(talking, hour):
  return (talking and (hour < 7 or hour > 20))

print(parrot_trouble(True,6))
```

```python
def pos_neg(a, b, negative):
  if negative:
    return (a < 0 and b < 0)
  else:
    return ((a < 0 and b > 0) or (a > 0 and b < 0))

print(pos_neg(1, -1, False))
```

```python
def front_back(str):
  if len(str) <= 1:
    return str
  
  mid = str[1:len(str)-1]  # can be written as str[1:-1]
  
  # last + mid + first
  return str[len(str)-1] + mid + str[0]

print(front_back('code'))
```

```python
def sum_double(a, b):
  # Store the sum in a local variable
  sum = a + b
  
  # Double it if a and b are the same
  if a == b:
    sum = sum * 2
  return sum


print(sum_double(1, 2))
```

```python
def makes10(a, b):
  return (a == 10 or b == 10 or a+b == 10)

print(makes10(9, 10))
```

```python
def not_string(str):
  if len(str) >= 3 and str[:3] == "not":
    return str
  return "not " + str

print(not_string('candy'))
```

```python
def front3(str):
  # Figure the end of the front
  front_end = 3
  if len(str) < front_end:
    front_end = len(str)
  front = str[:front_end]
  return front + front + front

print(front3('Chocolate'))
```

```python
def string_times(str, n):
  result = ""
  for i in range(n):  # range(n) is [0, 1, 2, .... n-1]
    result = result + str  # could use += here
  return result

print(string_times('Hi', 2))
```

```python
def string_splosion(str):
  result = ""
  # On each iteration, add the substring of the chars 0..i
  for i in range(len(str)):
    result = result + str[:i+1]
  return result

print(string_splosion('Code'))
```

```python
def array_front9(nums):
  # First figure the end for the loop
  end = len(nums)
  if end > 4:
    end = 4
  
  for i in range(end):  # loop over index [0, 1, 2, 3]
    if nums[i] == 9:
      return True
  return False

print(array_front9([1, 2, 9, 3, 4]))
```

```python
def front_times(str, n):
  front_len = 3
  if front_len > len(str):
    front_len = len(str)
  front = str[:front_len]
  
  result = ""
  for i in range(n):
    result = result + front
  return result

print(front_times('Chocolate', 2))
```

```python
def last2(str):
  # Screen out too-short string case.
  if len(str) < 2:
    return 0
  
  # last 2 chars, can be written as str[-2:]
  last2 = str[len(str)-2:]
  count = 0
  
  # Check each substring length 2 starting at i
  for i in range(len(str)-2):
    sub = str[i:i+2]
    if sub == last2:
      count = count + 1

  return count

print(last2('hixxhi'))
```

```python
def array123(nums):
  # Note: iterate with length-2, so can use i+1 and i+2 in the loop
  for i in range(len(nums)-2):
    if nums[i]==1 and nums[i+1]==2 and nums[i+2]==3:
      return True
  return False

print(array123([1, 1, 2, 4, 1]))
```

```python
def string_bits(str):
  result = ""
  # Many ways to do this. This uses the standard loop of i on every char,
  # and inside the loop skips the odd index values.
  for i in range(len(str)):
    if i % 2 == 0:
      result = result + str[i]
  return result

print(string_bits('Hello'))
```

```python
def array_count9(nums):
  count = 0
  # Standard loop to look at each value
  for num in nums:
    if num == 9:
      count = count + 1

  return count

print(array_count9([1, 9, 9, 3, 9]))
```

```python
def string_match(a, b):
  # Figure which string is shorter.
  shorter = min(len(a), len(b))
  count = 0
  
  # Loop i over every substring starting spot.
  # Use length-1 here, so can use char str[i+1] in the loop
  for i in range(shorter-1):
    a_sub = a[i:i+2]
    b_sub = b[i:i+2]
    if a_sub == b_sub:
      count = count + 1

  return count

print(string_match('xxcaazz', 'xxbaaz'))
```

```python
def double_char(s):
    return ''.join([c*2 for c in s])
print(double_char('The'))
print(double_char('AAbb'))
print(double_char('Hi-There'))
```

```python
def count_code(s):
    count = 0
    for i in range(len(s)-3):
        if s[i:i+2] == 'co' and s[i+3] == 'e':
            count += 1
    return count
print(count_code('aaacodebbb'))
print(count_code('codexxcode'))
print(count_code('cozexxcope'))
```

```python
def count_hi(s):
    return s.count('hi')
print(count_hi('abc hi ho'))
print(count_hi('ABChi hi'))
print(count_hi('hihi'))
```

```python
def end_other(a, b):
    a = a.lower()
    b = b.lower()
    return a.endswith(b) or b.endswith(a)
print(end_other('Hiabc', 'abc'))
print(end_other('AbC', 'HiaBc'))
print(end_other('abc', 'abXabc'))
```

```python
def cat_dog(s):
    return s.count('cat') == s.count('dog')
print(cat_dog('catdog'))
print(cat_dog('catcat'))
print(cat_dog('1cat1cadodog'))
```

```python
def xyz_there(s):
    if s.startswith('xyz'):
        return True
    for i in range(1, len(s)-2):
        if s[i:i+3] == 'xyz' and s[i-1] != '.':
            return True
    return False
print(xyz_there('abcxyz'))
print(xyz_there('abc.xyz'))
print(xyz_there('xyz.abc'))
```

```python
def count_evens(nums):
    return sum(1 for x in nums if x % 2 == 0)
print(count_evens([2, 1, 2, 3, 4]))
print(count_evens([2, 2, 0]))
print(count_evens([1, 3, 5]))
```

```python
def sum13(nums):
    total = 0
    skip = False
    for n in nums:
        if n == 13:
            skip = True
            continue
        if skip:
            skip = False
            continue
        total += n
    return total
print(sum13([1, 2, 2, 1]))
print(sum13([1, 1]))
print(sum13([1, 2, 2, 1, 13]))
```

```python
def big_diff(nums):
    return max(nums) - min(nums)
print(big_diff([10, 3, 5, 6]))
print(big_diff([7, 2, 10, 9]))
print(big_diff([2, 10, 7, 2]))
```

```python
def sum67(nums):
    total = 0
    in_block = False
    for n in nums:
        if n == 6:
            in_block = True
            continue
        if in_block and n == 7:
            in_block = False
            continue
        if not in_block:
            total += n
    return total
print(sum67([1, 2, 2]))
print(sum67([1, 2, 2, 6, 99, 99, 7]))
print(sum67([1, 1, 6, 7, 2]))
```

```python
def centered_average(nums):
    nums_sorted = sorted(nums)
    total = sum(nums_sorted[1:-1])
    return total // (len(nums) - 2)
print(centered_average([1, 2, 3, 4, 100]))
print(centered_average([1, 1, 5, 5, 10, 8, 7]))
print(centered_average([-10, -4, -2, -4, -2, 0]))
```
