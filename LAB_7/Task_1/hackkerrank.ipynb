```python
def solve(s):
    result = ""
    new_word = True

    for c in s:
        if new_word and c.isalpha():
            result += c.upper()
            new_word = False
        else:
            result += c
            if c == " ":
                new_word = True
            else:
                new_word = False

    return result

s = input()
print(solve(s))
```

```python
n = int(input())

students = {}

for _ in range(n):
    data = input().split()
    name = data[0]
    marks = list(map(float, data[1:]))
    students[name] = marks

query_name = input()

marks = students[query_name]
avg = sum(marks) / len(marks)

print(f"{avg:.2f}")
```

```python
n = int(input())

lst = []

for _ in range(n):
    cmd = input().split()

    if cmd[0] == "insert":
        lst.insert(int(cmd[1]), int(cmd[2]))
    elif cmd[0] == "print":
        print(lst)
    elif cmd[0] == "remove":
        lst.remove(int(cmd[1]))
    elif cmd[0] == "append":
        lst.append(int(cmd[1]))
    elif cmd[0] == "sort":
        lst.sort()
    elif cmd[0] == "pop":
        lst.pop()
    elif cmd[0] == "reverse":
        lst.reverse()
```

```python
def swap_case(s):
    return s.swapcase()

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```

```python
def print_formatted(number):
    w = len(bin(number)[2:])
    
    for i in range(1, number+1):
        print(
            str(i).rjust(w),
            oct(i)[2:].rjust(w),
            hex(i)[2:].upper().rjust(w),
            bin(i)[2:].rjust(w)
        )

n = int(input())
print_formatted(n)
```

```python
n = int(input())
s = set(map(int, input().split()))
m = int(input())

for _ in range(m):
    cmd = input().split()
    
    if cmd[0] == "pop":
        s.pop()
    elif cmd[0] == "remove":
        s.remove(int(cmd[1]))
    elif cmd[0] == "discard":
        s.discard(int(cmd[1]))

print(sum(s))
```
